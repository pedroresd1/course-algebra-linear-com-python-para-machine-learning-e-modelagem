## Tipos de Transformações

### Homotetia



Segmentos homólogos. Proporções iguais.

image.png

```python
#Vetor vetor[X,Y]
vetor = np.array([[2]
                   ,[1]])
vetor
```

```python
# Constante
k = 2
k
```

```python
# Matriz de transformacao
T = np.array([[k,0],
              [0,k]])
T
```

```python
# Coordenadas do Vetor
x = [vetor[0], 0]
y = [0, vetor[1]]
```

```python
plt.plot(x, y, color='blue');
```

```python
#Transformacao Hometetia (Aumentando o vetor 2x o seu tamanho)
matrizT = np.dot(T, vetor)
matrizT
```

```python
# Coordenadas da Transformacao Linear
xT = [matrizT[0], 0]
yT = [0, matrizT[1]]
```

```python
plt.plot(x, y, color='blue');
plt.plot(xT, yT, color='red');
plt.show();
```

### Translação

Movimento que mantém todas as suas propriedades geométricas (tamanho, ângulo e distâncias).

image.png

```python
# Vetor Original
vetorT = np.array([[3,2],
                    [2,5]])
vetorT
```

```python
# Coordenadas X e Y
x = [vetorT[0][0],vetorT[0][1]]
y = [vetorT[1][0],vetorT[1][1]]
```

```python
plt.plot(x, y, color='blue');
```

```python
# Matriz de Transformacao
T = np.array([[4,1],
              [4,1]])
```

```python
#Calculo Transformacao
matrizT = vetorT + T
matrizT
```

```python
# Coordenadas do Vetor de Transformacao XT e YT
xT = [matrizT[0][0],matrizT[1][0]]
yT = [matrizT[0][1],matrizT[1][1]]
```

```python
plt.plot(x, y, color='blue');
plt.plot(xT, yT, color='red');
plt.show();
```
